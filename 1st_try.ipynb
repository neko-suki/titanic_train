```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import svm
%matplotlib inline

train_data = pd.read_csv("train.csv")
```

```python
train_data.Age = train_data.Age.fillna(train_data.Age.mean())
train_data = train_data.replace('male', 0).replace('female', 1).replace('C', 0).replace('Q', 1).replace('S', 2)
train_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
train_data.dropna()
train_data.head(10)

male = train_data.drop(train_data.index[train_data.Sex==1])
male.drop(['Sex'], axis=1, inplace=True)

female = train_data.drop(train_data.index[train_data.Sex==0])
female.drop(['Sex'], axis=1, inplace=True)
female.Embarked = female.Embarked.fillna(2.0)
```

```python
male_sample_x = male.drop(['Survived'], axis=1)
male_sample_y = male.drop(["Pclass", "Age", "SibSp", "Parch", "Fare", "Embarked"], axis=1)
```

```python
male_sample_x
```

```python
male_sample_y
```

```python
female_sample_x = female.drop(['Survived'], axis=1)
female_sample_y = female.drop(["Pclass", "Age", "SibSp", "Parch", "Fare", "Embarked"], axis=1)
```

```python
female_sample_x
```

```python
male_svm = svm.SVC()
male_svm.fit(male_sample_x, male_sample_y.values.ravel())
```

```python
female_svm = svm.SVC()
female_svm.fit(female_sample_x, female_sample_y.values.ravel())
```

```python
test_data = pd.read_csv("test.csv")
test_data.Age = test_data.Age.fillna(test_data.Age.mean())
test_data = test_data.replace('male', 0).replace('female', 1).replace('C', 0).replace('Q', 1).replace('S', 2)
test_data.Embarked = test_data.Embarked.fillna(2.0)
test_data.Fare = test_data.Fare.fillna(0) # to be dead


ids = test_data['PassengerId']
sex = test_data['Sex']
test_data.drop(['PassengerId', 'Sex', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
test_data.head()
```

```python
output = pd.DataFrame(columns=["PassengerId","Survived"])


for i in range(len(ids)) :
    row = test_data[i:i+1]
    predicted = None
    if sex[i] == 0 : # male
        predicted = male_svm.predict(row)
    elif sex[i] == 1 : # female
        predicted = female_svm.predict(row)
    
    insert = pd.Series([ids[i], predicted[0]], index=output.columns)
    output = output.append(insert, ignore_index=True)

#    df = pd.DataFrame([[ids[i], predicted[0]]], columns=["PassengerId","Survived"])
#    output = output.append(df, ignore_index=True)
#print(output)
```
